## Version 2: Playing against a computer
This is my 2nd version where two players can play tiktak toe

```python
import random

# Tic-Tac-Toe

# Create the game board
board = [" " for _ in range(9)]

# Function to print the game board
def print_board():
    print("---------")
    print("|", board[0], "|", board[1], "|", board[2], "|")
    print("---------")
    print("|", board[3], "|", board[4], "|", board[5], "|")
    print("---------")
    print("|", board[6], "|", board[7], "|", board[8], "|")
    print("---------")

# Function to check if the game is over
def game_over():
    # Check rows
    for i in range(0, 9, 3):
        if board[i] == board[i + 1] == board[i + 2] != " ":
            return True
    # Check columns
    for i in range(3):
        if board[i] == board[i + 3] == board[i + 6] != " ":
            return True
    # Check diagonals
    if board[0] == board[4] == board[8] != " " or board[2] == board[4] == board[6] != " ":
        return True
    # Check for a tie
    if " " not in board:
        return True
    return False

# Function to make the player's move
def make_player_move():
    position = int(input("Enter a position (1-9): ")) - 1
    if board[position] == " ":
        board[position] = "X"
    else:
        print("Invalid move. Try again.")
        make_player_move()

# Function to make the computer's move
def make_computer_move():
    available_moves = [i for i in range(9) if board[i] == " "]
    if available_moves:
        position = random.choice(available_moves)
        board[position] = "O"

# Function to play the game
def play_game():
    current_player = "X"
    while not game_over():
        print_board()
        if current_player == "X":
            make_player_move()
        else:
            make_computer_move()
        current_player = "O" if current_player == "X" else "X"
    print_board()
    if " " not in board:
        print("It's a tie!")
    else:
        winner = "Player" if current_player == "X" else "Computer"
        print(winner, "wins!")

# Start the game
play_game()
```

```python
import random

# Tic-Tac-Toe

# Create the game board
board = [" " for _ in range(9)]

# Function to print the game board
def print_board():
    print("---------")
    print("|", board[0], "|", board[1], "|", board[2], "|")
    print("---------")
    print("|", board[3], "|", board[4], "|", board[5], "|")
    print("---------")
    print("|", board[6], "|", board[7], "|", board[8], "|")
    print("---------")

# Function to check if the game is over
def game_over():
    # Check rows
    for i in range(0, 9, 3):
        if board[i] == board[i + 1] == board[i + 2] != " ":
            return True
    # Check columns
    for i in range(3):
        if board[i] == board[i + 3] == board[i + 6] != " ":
            return True
    # Check diagonals
    if board[0] == board[4] == board[8] != " " or board[2] == board[4] == board[6] != " ":
        return True
    # Check for a tie
    if " " not in board:
        return True
    return False

# Function to make the player's move
def make_player_move(player_symbol):
    position = int(input("Enter a position (1-9): ")) - 1
    if board[position] == " ":
        board[position] = player_symbol
    else:
        print("Invalid move. Try again.")
        make_player_move(player_symbol)

# Function to make the computer's move
def make_computer_move(computer_symbol):
    available_moves = [i for i in range(9) if board[i] == " "]
    if available_moves:
        position = random.choice(available_moves)
        board[position] = computer_symbol

# Function to play the game
def play_game():
    player_symbol = input("Choose your symbol (X/O): ").upper()
    if player_symbol == "X":
        computer_symbol = "O"
    else:
        computer_symbol = "X"

    current_player = player_symbol
    while not game_over():
        print_board()
        if current_player == player_symbol:
            make_player_move(player_symbol)
        else:
            make_computer_move(computer_symbol)
        current_player = computer_symbol if current_player == player_symbol else player_symbol
    print_board()
    if " " not in board:
        print("It's a tie!")
    else:
        winner = "Player" if current_player == player_symbol else "Computer"
        print(winner, "wins!")

# Start the game
play_game()
```
